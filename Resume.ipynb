```python
# prompt: Extract Key Information:
# Use the Resume column to extract the text content (skills, keywords, descriptions).
# Ensure the text is tokenized, cleaned, and free of duplicates or irrelevant data and numbers.

import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def clean_resume_text(text):
    text = re.sub('[^a-zA-Z]', ' ', text)
    text = text.lower()
    text = nltk.word_tokenize(text)
    text = [lemmatizer.lemmatize(word) for word in text if not word in stop_words]
    text = [word for word in text if len(word) > 1]
    return " ".join(text)


df['Cleaned_Resume'] = df['Resume'].apply(lambda x: clean_resume_text(x))
df['Cleaned_Resume'].head() # Display the first few cleaned resumes

# Further processing (optional):
# Convert to a list of unique words if you want to avoid repeated words within a resume
# Create a vocabulary of unique words across all resumes.

# Example of creating a vocabulary:
all_words = []
for index in df.index:
  all_words.extend(df.loc[index, 'Cleaned_Resume'].split()) #Split the string on space
vocabulary = set(all_words)
print(f"Vocabulary size: {len(vocabulary)}")
```

```python
import os
from pinecone import Pinecone, ServerlessSpec
from google.colab import userdata


# Initialize Pinecone
pinecone_api_key = userdata.get("PINECONE_API_KEY")  # Replace with your API key

pc = Pinecone(api_key=pinecone_api_key)

# Define index name and configuration
index_name = "resume-dataset-index"

# Delete the existing index if it exists
if index_name in [idx["name"] for idx in pc.list_indexes()]:
    pc.delete_index(index_name)

# Create the index with the correct dimension
pc.create_index(
    name=index_name,
    dimension=768,  # Match Google Generative AI embedding dimension
    metric="cosine",
    spec=ServerlessSpec(cloud="aws", region="us-east-1")
)

index = pc.Index(index_name)
```

```python
def split_text(text, chunk_size=1000):
    """
    Split text into smaller chunks of a specified size.
    """
    return [text[i:i + chunk_size] for i in range(0, len(text), chunk_size)]
```

```python
from langchain_core.documents import Document

chunk_size = 1000  # Define the size of each chunk
documents = []

for i, row in df.iterrows():
    resume_chunks = split_text(row["Cleaned_Resume"], chunk_size=chunk_size)
    for chunk_id, chunk in enumerate(resume_chunks):
        documents.append(
            Document(
                page_content=chunk,  # Store the resume content in chunks
                metadata={
                    "Category": row["Category"],
                    "ResumeID": f"id_{i}",  # Unique ID for the full resume
                    "ChunkID": chunk_id    # ID for each chunk
                }
            )
        )
```

```python
import json

def validate_metadata(metadata):
    """
    Validate that metadata size is within Pinecone's 40 KB limit.
    """
    metadata_size = len(json.dumps(metadata).encode('utf-8'))
    if metadata_size > 40960:  # Pinecone metadata size limit
        raise ValueError(f"Metadata size {metadata_size} exceeds the limit of 40960 bytes")

# Validate metadata for all documents
for doc in documents:
    validate_metadata(doc.metadata)
```

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_pinecone import PineconeVectorStore

# Initialize embeddings
embeddings = GoogleGenerativeAIEmbeddings(
    model="models/embedding-001",
    google_api_key=userdata.get("GOOGLE_API_KEY")
)

# Initialize Pinecone index
from pinecone import Pinecone, ServerlessSpec

pc = Pinecone(api_key=userdata.get("PINECONE_API_KEY"))
index_name = "resume-dataset-index"
if index_name not in [idx["name"] for idx in pc.list_indexes()]:
    pc.create_index(
        name=index_name,
        dimension=1536,
        metric="cosine",
        spec=ServerlessSpec(cloud="aws", region="us-east-1")
    )
index = pc.Index(index_name)

# Initialize vector store
vector_store = PineconeVectorStore(index=index, embedding=embeddings)

# Generate embeddings and add to vector store
ids = [f"{doc.metadata['ResumeID']}_chunk_{doc.metadata['ChunkID']}" for doc in documents]
vector_store.add_documents(documents=documents, ids=ids)
```

```python
query = "Looking for a Data Scientist with Python and machine learning expertise."
results = vector_store.similarity_search(query=query, k=3)

for result in results:
    print(f"Category: {result.metadata['Category']}")
    print(f"ChunkID: {result.metadata['ChunkID']}")
    print(f"Resume Chunk Content: {result.page_content}\n")
```
